**Lane:** Discoverability & Ranking Decline (Content Decay)

**Why:** Identifying which pages are losing organic search visibility before their traffic bottoms out is a high-ROI business problem. ## 2. The question: decision, action, cost of a wrong call

* **The Action:** The content team assigns an editor to update, re-optimize, or expand the flagged pages to recover lost search visibility.
* **The Cost of a wrong call:** False positives waste expensive editorial hours on pages that are naturally fluctuating but technically fine. False negatives result in high-value pages slowly bleeding impressions and dropping out of the top search results entirely unnoticed.

## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

```python
import pandas as pd
import glob
import os

# Locate the starter CSV in the data folder (going up two directories from work/notebooks)
data_files = glob.glob("../../data/*.csv")

if not data_files:
    print("Error: Could not find the starter CSV in the data folder. Please check your path.")
else:
    # Load the first available CSV from the starter data
    file_path = data_files[0]
    df = pd.read_csv(file_path)
    
    total_records = len(df)
    
    # Safely extract metrics assuming standard search console column naming
    impressions_col = next((col for col in df.columns if 'impression' in col.lower()), None)
    ctr_col = next((col for col in df.columns if 'ctr' in col.lower()), None)
    
    print("--- Data Validation for Content Decay Lane ---")
    print(f"Total search records available for analysis: {total_records:,}")
    
    if impressions_col:
        avg_impressions = df[impressions_col].mean()
        print(f"Average impressions across dataset: {avg_impressions:,.1f}")
        
    if ctr_col:
        low_ctr_count = len(df[df[ctr_col] < 0.02])
        print(f"Records with critically low CTR (< 2%): {low_ctr_count:,} pages at risk")
```
