## 1. My lane as an ML task (type)

*Classification, clustering, ranking, or scoring — which one, and why?*

**Lane:** Content Decay / Refresh Prioritization.

**Task Type:** Binary Classification. 
The machine learning system will categorize existing pages into one of two distinct classes: `1` (Requires Immediate Refresh) or `0` (Stable/Healthy). This translates a continuous traffic problem into a clear, actionable flag for the editorial team.

## 2. Target or proxy

*What would you predict? Where does that label come from — observed outcome or a defined rule?*

**The Proxy Target:** Since we cannot directly measure Google's internal ranking scores or algorithm shifts, we must use a proxy target derived from our own search data. 

The target will be a binary label: `is_decaying`. We will define this as `1` if a page's average daily search impressions have dropped by more than 20% over the last 30 days compared to the previous 30-day period, and `0` otherwise.

## 3. Success metric

*One metric you can defend. What number means 'good'?*

**Success Metric:** Precision. 

In this business context, the cost of a false positive is highly expensive—assigning an editor to update, rewrite, and QA a page that is naturally fluctuating but ultimately fine wastes hours of payroll. Therefore, we must optimize for Precision; when the model flags a page as decaying, we need to be highly confident that it genuinely needs human intervention.

## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*

A fixed rule (e.g., "rewrite all pages older than 180 days" or "flag pages when CTR drops below 2%") is too rigid and highly inefficient. It ignores the nuanced reality of search dynamics: a 2-year-old, high-quality definitive guide might still be gaining traffic, while a 3-month-old trending article might be decaying rapidly. ML beats fixed heuristics here because it can learn the complex, non-linear interactions between content age, historical CTR trajectories, and impression volatility, allowing us to identify structural decay rather than just normal seasonal dips.
